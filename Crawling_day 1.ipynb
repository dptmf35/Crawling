## Beautiful Soup

```python
## 가지고 온 페이지 정보 중 내가 실제 필요한 데이터만 추출할 때 사용
```

```python
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
naver_url = "https://www.naver.com/"
```

```python
res = req.get(naver_url)
```

```python
info = res.text
```

```python
naver_res = bs(info, 'lxml')
```

```python
list2= naver_res.find_all('li', class_ = "nav_item") 
#해당 태그 모두 찾기 find_all(찾을 태그명,태그의 클래스)
# >>>데이터를 리스트 형태로 반환
#아이디값 찾기 find
```

```python
list2[0]
```

```python
list2[0].text
```

```python
s = list2[0].text
s.strip() # 문자열의 양 끝에 존재하는 공백과 \n을 제거
```

```python
for i in list2 :
    print(i.text.strip())
```

## 네이버에서 코로나 검색 후 뉴스 제목만 크롤링

```python
# 0. 크롤링 시 import문
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
# 1. url 지정
url = "https://search.naver.com/search.naver?where=nexearch&sm=tab_jum&query=%EC%BD%94%EB%A1%9C%EB%82%98"
```

```python
# 2. url 불러오기
res = req.get(url) #요청하여 응답 받기
info = res.text  #url 내용 추출
```

```python
# 3. beautiful soup로 필요한 데이터만 추출
c_res = bs(info, 'lxml') #html코드 간편화
arr= c_res.find_all('a', class_ = "_sp_each_title") #태그,클래스 이용 다중 데이터 찾기
```

```python
for i in arr :
    print(i.text.strip())
```
