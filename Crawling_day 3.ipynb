```python
## 페이지 번호 바꿔서 크롤링 1 ~ 10 page
```

```python
url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189537&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="
```

```python
for i in range(1, 11) :
    url = url + str(i)
    res = req.get(url)
    review = bs(res.text,'lxml')
    m_review = review.select('div.score_reple > p')
    icon = review.select('span.ico_viewer')
    spo = review.select('span.text_spo')
    for a in icon :
        a.extract()
    for b in spo :
        b.extract()
    for j in m_review :
        print("- " + j.text.strip())
```

```python
r_list = []
for i in range(1, 11) :
    url = url + str(i)
    res = req.get(url)
    review = bs(res.text,'lxml')
    m_review = review.select('div.score_reple > p')
    icon = review.select('span.ico_viewer')
    spo = review.select('span.text_spo')
    
    for a in icon :
        a.extract()
    
    for b in spo :
        b.extract()
    
    for j in range(len(m_review)) :
        r_list.append(m_review[j].text.strip())

print(r_list)
    
```
