### 타임으로 실행 시간 딜레이 주기

```python
import time
```

### 예외처리 : Try 에서 에러날 시 Except문 실행

```python
try :
    for i in range(50) :
        p_click.click()
        time.sleep(1) ##1초에 한번씩 실행
except :
    print(i) ##예외처리로 클릭할게 없어도 에러 안나게 해줌
```

```python
source = driver.page_source ##현재 페이지 주소
## bs(driver.page_source,'lxml')
```

```python
b_source = bs(source,'lxml')
hansot_menu = b_source.select('div.item-text > h4')
hansot_menu
```

```python
for i in hansot_menu :
    print(i.text.strip())
```

```python
hansot_price = b_source.select('span.blind ~ strong')
hansot_price
```

```python
for i in hansot_price :
    print(i.text.strip())
```

```python
import pandas as pd
```

```python
menu_list = []
price_list = []
num_list = []
num = 1
for i in range(1,len(hansot_menu)+1) :
    num_list.append(i)
    num += 1
for i in hansot_menu :
    menu_list.append(i.text.strip())

for i in hansot_price :
    price_list.append(i.text.strip())
```

```python
len(price_list)
```

```python
num_list
```

```python
hansot_info = {'num' : num_list, 'menu' : menu_list, 'price' : price_list}

hansot_data = pd.DataFrame(hansot_info, )
```

## 유튭 영상 제목 가져오기

```python
url = "https://www.youtube.com/results?search_query=%ED%94%84%EB%A0%88%EC%9D%B4"
```

```python
driver = wb.Chrome()

driver.get(url)
```

```python
soup = bs(driver.page_source,'lxml')
```

```python
youtube = soup.select('a#video-title > yt-formatted-string')
for i in youtube :
    print(i.text)
```

```python
len(youtube)
```

```python
from tqdm import tqdm_notebook as tq
```

```python
body = driver.find_element_by_tag_name('body') #바디
body.send_keys(Keys.PAGE_DOWN)
```

```python
for i in tq(range(10)):
    body.send_keys(Keys.PAGE_DOWN) #페이지 다운 키로 스크롤 내리기
    time.sleep(1)
```

```python
body = driver.find_element_by_css_selector('body') #바디에 보내주기
for i in tq(range(10)):
    body.send_keys(Keys.PAGE_DOWN) #페이지 다운 키로 스크롤 내리기
    time.sleep(1)
source = driver.page_source
soup2 = bs(source,'lxml')
youtube = soup2.select('a#video-title > yt-formatted-string')
for i in youtube :
    print(i.text)
```

## 맘스터치 메뉴

```python
url2 = "http://www.momstouch.co.kr/sub/menu/menu_list.html"
```

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys 
```

```python
from bs4 import BeautifulSoup as bs
```

```python
driver = wb.Chrome()
```

```python
driver.get(url2)
```

```python
source = driver.page_source
```

```python
menu_page = bs(source,'lxml')
```

```python
menu_page.select('span.title')
```

```python
menu_list = []

for i in range(3) :    
    menu_page = bs(source,'lxml')
    menu = menu_page.select('span.title')
    for j in menu :
        menu_list.append(j.text)
    btn_click = driver.find_element_by_class_name('page_next')
    btn_click.click()
    source = driver.page_source
    time.sleep(1)
```
