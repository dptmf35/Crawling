```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
```

```python
import time
```

```python
url = "https://www.starbucks.co.kr/store/store_map.do"
```

```python
driver = wb.Chrome()
```

```python
driver.get(url)
```

```python
#지역 검색 클릭
loc_btn = driver.find_element_by_class_name('loca_search')
loc_btn.click()
```

```python
#xpath로 지정할 태그 찾기 (copy xpath)
g_btn_path = driver.find_element_by_xpath('//*[@id="container"]/div/form/fieldset/div/section/article[1]/article/article[2]/div[1]/div[2]/ul/li[3]/a')
```

```python
g_btn_path.click()
```

```python
#광주 클릭
g_btn = driver.find_element_by_css_selector('ul.sido_arae_box > li + li + li')
g_btn.click()
```

```python
# 전체 클릭
all_btn = driver.find_element_by_class_name('set_gugun_cd_btn')
all_btn.click()
```

```python
# 매장명 수집
source = driver.page_source
b_source = bs(source,'lxml')
store_name = b_source.select('ul.quickSearchResultBoxSidoGugun > li.quickResultLstCon > strong')
store_name[0].text
```

```python
store_name
```

```python
name_list = []
for i in  store_name :
    name_list.append(i.text.strip())
print(name_list)
```

```python
len(name_list)
```

```python
#주소 수집

store_add = b_source.select('ul.quickSearchResultBoxSidoGugun > li.quickResultLstCon > p.result_details')
```

```python
store_add[0].text[-9:]
```

```python
#전화번호 
tel_list = []
for i in store_add :
    tel_list.append(i.text[-9:])
print(tel_list)
```

```python
s = '광주광역시 서구 풍암2로 8 (풍암동) 한아름빌딩1522-3232'
s.strip(s[-9:])
```

```python
add_list = []
for i in store_add :
    add_list.append(i.text.strip(i.text[-9:]))
    
print(add_list)
```

```python
rank = []
for i in range(1,len(name_list)+1) : 
    rank.append(i)
print(rank)
```

```python
import pandas as pd
```

```python
starbucks_info = {"rank" : rank, "name" : name_list , "add" : add_list, "tel" : tel_list}
```

```python
data = pd.DataFrame(starbucks_info)
data
```

```python
data.set_index('rank', inplace = True)
data
```

# 이미지 저장 크롤링

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
```

```python
import os #폴더 생성 또는 삭제
from urllib.request import urlretrieve #이미지 경로를 파일로 저장
```

```python
url = "https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EA%B3%A0%EC%96%91%EC%9D%B4"
```

```python
#저장할 폴더 만들기
if not os.path.isdir('./떼껄룩') : 
    print("폴더생성")
    os.mkdir('./떼껄룩')
```

```python
driver = wb.Chrome()

driver.get(url)
```

```python
soup = bs(driver.page_source,'lxml')
```

```python
img = soup.select('img._img')
```

```python
img[0]['src']
```

```python
img_list = []
for i in img :
    img_list.append(i['src'])
```

```python
len(img_list)
```

```python
import time
from tqdm import tqdm_notebook as tq
```

```python
for i in tq(range(len(img_list))) :
    
    urlretrieve(img_list[i], './떼껄룩/' + '떼껄룩' + str(i+1) + '.jpg')
    
    time.sleep(1)
```

```python
url = "https://www.starbucks.co.kr/menu/food_list.do"
```

```python
name = soup.select('li.menuDataSet > dl > dd')
name
```

```python
len(name)
```

```python
n_list = []
for i in name :
    n_list.append(i.text)
print(n_list)
```

```python
if not os.path.isdir('./스타벅스') : 
    print("폴더생성")
    os.mkdir('./스타벅스')
```

```python
for i,j in tq(zip(range(len(img_list)),n_list)) :
    
    urlretrieve(img_list[i], './스타벅스/' + j + '.jpg')
    
    time.sleep(1)
```
